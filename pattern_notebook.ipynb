```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
# Seems like working and also clean.
a = [1, 2, 3]
b = [4, 5, 6]
c = [7, 8, 9]

result = []

for i, x in enumerate(a):
    b_iter = b if i % 2 == 0 else b[::-1]
    for j, y in enumerate(b_iter):
        # Alternate direction of c depending on (i + j) % 2
        c_iter = c if (i + j) % 2 == 0 else c[::-1]
        for z in c_iter:
            result.append((x, y, z))

for res in result:
    print(res)
```

```python
# rcParams for matplotlib
plt.rcParams.update(
    {
        # Math text
        "mathtext.default": "regular",
        # Figure
        "figure.figsize": [4, 4],  # or [4, 6.4]
        "figure.dpi": 300,
        # Plot (lines and markers)
        "lines.linewidth": 3,
        "lines.markersize": 8,
        # Axes
        "axes.titlesize": 16,
        "axes.titleweight": "bold",
        "axes.labelsize": 16,
        "axes.labelweight": "bold",
        "xtick.labelsize": 15,
        "ytick.labelsize": 15,
        "xtick.direction": "in",
        "ytick.direction": "in",
    }
)

# Plotting the code in previous cell and showing the direction of zigzag pattern
def generate_zigzag_pattern(a, b, c):
    result = []
    for i, x in enumerate(a):
        b_iter = b if i % 2 == 0 else b[::-1]
        for j, y in enumerate(b_iter):
            # Alternate direction of c depending on (i + j) % 2
            c_iter = c if (i + j) % 2 == 0 else c[::-1]
            for z in c_iter:
                result.append((x, y, z))
                
    return result


# Example input
a = [1, 2, 3]
b = [4, 5, 6]
c = [7, 8, 9]

# Generate pattern
points = generate_zigzag_pattern(a, b, c)

# Extract x, y, z coordinates
x_vals, y_vals, z_vals = zip(*points)

# Create 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")

# Plot the points
ax.scatter(x_vals, y_vals, z_vals, color="blue", s=50, label="Points")

# Draw arrows between consecutive points to show direction
for i in range(len(points) - 1):
    x0, y0, z0 = points[i]
    x1, y1, z1 = points[i + 1]
    dx, dy, dz = x1 - x0, y1 - y0, z1 - z0
    ax.quiver(x0, y0, z0, dx, dy, dz, color="red", arrow_length_ratio=0.1)

# Set rcParams
# Publication quality figure settings
import matplotlib.pyplot as plt


# Label axes
ax.set_xlabel("Decision_variable_1")
ax.set_ylabel("Decision_variable_2")
ax.set_zlabel("Decision_variable_3")
ax.set_title("Progression of decision variables in decision space")
plt.legend()
plt.tight_layout()
plt.savefig("zigzag_pattern.png", dpi=300)
plt.show()
```

```python
# Just a regular for loop
# Plotting the code in previous cell and showing the direction of zigzag pattern
def generate_zigzag_pattern(a, b, c):
    result = []
    for i, x in enumerate(a):
        for j, y in enumerate(b):
            for z in c:
                result.append((x, y, z))
    return result


# Example input
a = [1, 2, 3]
b = [4, 5, 6]
c = [7, 8, 9]

# Generate pattern
points = generate_zigzag_pattern(a, b, c)

# Extract x, y, z coordinates
x_vals, y_vals, z_vals = zip(*points)

# Create 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")

# Plot the points
ax.scatter(x_vals, y_vals, z_vals, color="blue", s=50, label="Points")

# Draw arrows between consecutive points to show direction
for i in range(len(points) - 1):
    x0, y0, z0 = points[i]
    x1, y1, z1 = points[i + 1]
    dx, dy, dz = x1 - x0, y1 - y0, z1 - z0
    ax.quiver(x0, y0, z0, dx, dy, dz, color="red", arrow_length_ratio=0.1)

# Label axes
ax.set_xlabel("a")
ax.set_ylabel("b")
ax.set_zlabel("c")
ax.set_title("Zigzag Pattern Direction")
plt.legend()
plt.tight_layout()
plt.show()
```
